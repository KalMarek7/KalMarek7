```
# Start the server in the background
import subprocess
import threading
import time

def run_ollama_serve():
    subprocess.Popen(["ollama", "serve"])

thread = threading.Thread(target=run_ollama_serve)
thread.start()

# Wait a few seconds for the server to wake up
time.sleep(5)
print("🚀 Ollama server is up and running!")
```

```
import requests

# Pull the model (this will show a progress bar and finish)
!ollama pull gemma3:4b

# Define your extraction function
def test_extraction(text):
    url = "http://localhost:11434/api/generate"
    payload = {
        "model": "gemma3:4b",
        "prompt": f"Extract JSON with keys 'name' and 'price' from this text: {text}",
        "format": "json",
        "stream": False
    }
    response = requests.post(url, json=payload)
    return response.json()['response']

# Run the test
my_data = "I bought a cool gadget for 45 dollars yesterday."
print(test_extraction(my_data))
```
